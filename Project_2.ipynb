```python
import random
#encryption function defined below 
'''
def encrypt(plaintext):
    ciphertext= ""
    for i in range(0, len(plaintext)):
        print(i)
        if i % 2 ==0:
            plain_ascii = ord(plaintext[i])
            cipher_ascii=plain_ascii+1
            ciphertext += chr(cipher_ascii)
        else:
            plain_ascii = ord (plaintext[i])
            cipher_ascii=plain_ascii-1
            ciphertext += chr(cipher_ascii)

    return ciphertext

'''
#decryption function defined below

def decrypt(plaintext):
    ciphertext= ""
    for i in range(0, len(plaintext)):
        print(i)
        if i % 2 ==0:
            plain_ascii = ord(plaintext[i])
            cipher_ascii=plain_ascii-1
            ciphertext += chr(cipher_ascii)
        else:
            plain_ascii = ord (plaintext[i])
            cipher_ascii=plain_ascii+1
            ciphertext += chr(cipher_ascii)

    return ciphertext



words = ["WFIDSPQZ", "XHO' 'TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]

#function will choose random words from the list.

word = random.choice(words)
print("Word to guess: " + str(word))

guesses = ''
turns = 10

#we will use the while loop here.
#check if the turns are more than 0 then make a counter that starts with zero. For every character in random.choice words. 
#see if the character is in the input guess, then print the character. If the character is not found. then print a dash. 
#then increase the failed counter with one.

while turns > 0:
    count = 0
    for char in word:
        if char in guesses:
            print (char)
        else:
            print ("_")
            count += 1

    if count == 0:
        print ("You won")
        print(decrypt(word))
        break

    guess = input("guess a character:")
    guesses += guess

    if guess not in word:
        turns -= 1
        print ("Wrong")
        print ("You have", + turns, 'more guesses')

        if turns == 0:
            print("You Lose")
```
